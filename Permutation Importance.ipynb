```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("FIFA 2018 Statistics.csv")
```

```python
y = (df['Man of the Match']=='Yes')

# Filtering the numerical cols
num_cols = [i for i in df.columns if df[i].dtypes in [np.int64]]

X = df[num_cols]

trainX,valX,trainY,valY = train_test_split(X,y,random_state=1)

model = RandomForestClassifier(n_estimators=100,random_state=0).fit(trainX,trainY)
```

```python
import eli5
from eli5.sklearn import PermutationImportance

perm = PermutationImportance(model, random_state=1).fit(valX, valY)
eli5.show_weights(perm, feature_names = valX.columns.tolist())
```
